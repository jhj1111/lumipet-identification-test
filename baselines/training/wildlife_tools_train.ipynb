# Trainer

```python
import os
import shutil
import torch
from wildlife_tools.train import BasicTrainer

class CustomTrainer(BasicTrainer):
    def __init__(self, patience=5, save_path='/content/best_model.pth', backup_path=None, **kwargs):
        # 1. 부모 클래스(BasicTrainer)의 모든 설정 초기화
        super().__init__(**kwargs)

        # 2. 커스텀 변수 추가
        self.patience = patience
        self.save_path = save_path
        self.backup_path = backup_path
        self.best_loss = float('inf')
        self.patience_counter = 0

    def train(self):
        """부모 클래스의 train() 메서드를 오버라이딩하여 커스텀 루프 적용"""
        print(f"🚀 CustomTrainer 시작 (Patience: {self.patience})...")

        # DataLoader 생성 (부모 클래스와 동일한 방식)
        loader = torch.utils.data.DataLoader(
            self.dataset,
            batch_size=self.batch_size,
            num_workers=self.num_workers,
            shuffle=True,
        )

        for e in range(self.epochs):
            # 💡 핵심: 부모 클래스의 train_epoch()를 그대로 활용!
            # 내부적으로 순전파/역전파/스케줄러 업데이트/tqdm 프로그레스 바가 모두 알아서 돌아감
            epoch_data = self.train_epoch(loader)
            self.epoch += 1

            # 반환된 딕셔너리에서 평균 Loss 추출
            epoch_loss = epoch_data["train_loss_epoch_avg"]
            print(f"\n📊 Epoch {self.epoch} 요약 | Avg Loss: {epoch_loss:.4f}")

            # --- [Early Stopping & Best Model 저장 로직] ---
            if epoch_loss < self.best_loss:
                self.best_loss = epoch_loss
                self.patience_counter = 0

                # 💡 핵심: 부모 클래스의 강력한 save() 메서드 활용
                save_folder = os.path.dirname(self.save_path)
                save_file = os.path.basename(self.save_path)
                self.save(folder=save_folder, file_name=save_file)
                print(f"🌟 Best model saved! (Loss가 {self.best_loss:.4f} 로 갱신됨)")

                # 구글 드라이브 백업
                if self.backup_path:
                    shutil.copy(self.save_path, self.backup_path)
            else:
                self.patience_counter += 1
                print(f"⚠️ Loss 미개선 (Patience: {self.patience_counter}/{self.patience})")

                if self.patience_counter >= self.patience:
                    print(f"🛑 Early Stopping! 에폭 {self.epoch}에서 조기 종료합니다.")
                    break

            # 부모 클래스의 콜백 기능도 그대로 유지
            if self.epoch_callback:
                self.epoch_callback(trainer=self, epoch_data=epoch_data)
```

```python
pre_trained = False
if os.path.exists(gdrive_weight_path) :
  shutil.copy(gdrive_weight_path, colab_weight_path)
  pre_trained = True
  print(f"구글 드라이브에서 기존 가중치를 불러왔습니다: {gdrive_weight_path}")
```

```python
# 메모리 절약을 위해 Gradient Checkpointing 활성화
is_set_gradient_checkingpoint = True

def set_gradient_checkpointing(backbone, is_set_gradient_checkingpoint: bool):
  if hasattr(backbone, 'set_grad_checkpointing') and is_set_gradient_checkingpoint:
    global accumulation_steps
    backbone.set_grad_checkpointing(enable=True)
    accumulation_steps=int(32/batch_size)
    print("Gradient Checkpointing이 활성화되었습니다. (메모리 절약 모드)")
```

```python
from itertools import chain
import timm
import torch
import torchvision.transforms as T
from torch.optim import SGD

from wildlife_tools.data import WildlifeDataset
from wildlife_tools.train import ArcFaceLoss
from wildlife_datasets.splits import ClosedSetSplit

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)

# 2. Train / Test Split
splitter = ClosedSetSplit(ratio_train=ratio_train)
train_idx, test_idx = next(iter(splitter.split(df)))
train_df = df.loc[train_idx]

# 3. Transform 및 이미지 사이즈 설정
img_size = int(cfg[model_name]['img_size'])
transform = T.Compose([
    T.Resize((img_size, img_size)), # 고양이 식별은 얼굴/무늬가 안 잘리게 RandomCrop보다 Resize를 추천
    T.RandAugment(num_ops=2, magnitude=20),
    T.ToTensor(),
    T.Normalize(mean=mean, std=std),
])

# 4. WildlifeDataset 생성 (두 패키지의 연결 고리)
# train_df에 이미 파일 경로(path)와 식별자(identity) 컬럼이 들어있어.
dataset = WildlifeDataset(metadata=train_df, root=cat_data.root, transform=transform)

backbone = timm.create_model(str(cfg[model_name]['model_name']), num_classes=0, pretrained=True)

set_gradient_checkpointing(backbone, is_set_gradient_checkingpoint)

if pre_trained :
  backbone.load_state_dict(torch.load(colab_weight_path, map_location=device))
  print("기존 가중치가 모델에 성공적으로 로드되었습니다.")

with torch.no_grad():
    dummy_input = torch.randn(1, 3, img_size, img_size)
    embedding_size = backbone(dummy_input).shape[1]

# dataset.num_classes는 학습 데이터에 있는 고양이의 총 마리(ID) 수
objective = ArcFaceLoss(num_classes=dataset.num_classes, embedding_size=embedding_size, margin=0.5, scale=64)

# 6. Optimizer 및 Scheduler
params = chain(backbone.parameters(), objective.parameters())
optimizer = SGD(params=params, lr=lr, momentum=0.9)
min_lr = optimizer.defaults.get("lr") * 1e-3
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs, eta_min=min_lr)
```

# Train

```python
# 기존 BasicTrainer 대신 CustomTrainer 객체 생성
trainer = CustomTrainer(
    # --- 새로 추가된 파라미터 ---
    patience=5,
    save_path=colab_weight_path,     # '/content/cat_reid_weight.pth'
    backup_path=gdrive_weight_path,  # 구글 드라이브 경로
    # --- 부모 클래스(BasicTrainer) 파라미터 ---
    dataset=dataset,
    model=backbone,
    objective=objective,
    optimizer=optimizer,
    scheduler=scheduler,
    batch_size=batch_size,
    accumulation_steps=accumulation_steps,
    num_workers=2,
    epochs=epochs,
    device=device,
)
```

```python
trainer.train()
```
